We next look at the missing data. The code below shows that the missing data is encoded as `?` or `NaN`. We find the number of such values.

```python
# Check the unique values in the columns with missing data
print("Unique values in 'workclass':\n", X['workclass'].unique())
print("\nUnique values in 'occupation':\n", X['occupation'].unique())
print("\nUnique values in 'native-country':\n", X['native-country'].unique())

# Count the number of NaN values
nan_count = X['workclass'].isna().sum()

# Count the number of '?' values
question_mark_count = (X['workclass'] == '?').sum()

# Output the results
print(f"\nNumber of NaN values: {nan_count}")
print(f"Number of '?' values: {question_mark_count}")
```

```python
# Function to count NaN and '?' values
def count_missing_values(column):
    nan_count = column.isna().sum()  # Count NaN values
    question_mark_count = (column == '?').sum()  # Count '?' values
    return nan_count + question_mark_count  # Return total missing counts

# Count missing values per column
missing_counts = X.apply(count_missing_values)

# Display the overall missing values per column
print("\nMissing values per column:\n", missing_counts)  # Total missing values count

# Display missing values in target variable (if needed)
print("\nMissing values in target variable:\n", y.isnull().sum())  # Missing values count
```

We see that three variables have missing data and that these are all categorical variables. We look at the missingness in more detail, seeking to learn whether it is related to income.

We attempt to find the numbers of missing data according to income. However, inspecting the income variable reveals that data cleaning is required. This can be seen from the output of the code below.

```python
# View unique categories of the Income in y
unique_categories = y[y.columns[0]].unique()
print("Unique categories in the Income:")
print(unique_categories)
```

We merge the categories appropriately below, and view the missing values by income class. We also look for patterns in the missingness.

```python
# Map '?' to NaN in X
X.replace('?', np.nan, inplace=True)

# Merge the categories by replacing the values using .loc
y.loc[:, y.columns[0]] = y[y.columns[0]].replace({'<=50K.': '<=50K', '>50K.': '>50K'})

# Count missing values grouped by income class
missing_counts = X.isnull().groupby(y[y.columns[0]]).sum()

# Count total people in each income class
total_counts = y[y.columns[0]].value_counts()

# Combine missing counts and total counts into a single DataFrame
missing_summary = pd.DataFrame({
    'Missing Values': missing_counts.sum(axis=1),
    'Total Count': total_counts
})

# Display the missing values by income class along with total counts
print("\nMissing values by income class:\n", missing_summary)

# Display the breakdown of missing values for each variable
breakdown_counts = X.isnull().groupby(y[y.columns[0]]).sum()
print("\nBreakdown of missing values by variable and income class:\n", breakdown_counts)

# Step 1: Select only the relevant columns
selected_columns = ['workclass', 'occupation', 'native-country']
X_selected = X[selected_columns]

# Step 2: Create the heatmap to visualize missing data
plt.figure(figsize=(10, 6))
sns.heatmap(X_selected.isnull(), cbar=False, cmap='viridis', yticklabels=False)

# Add labels and title
plt.title('Missing Data in workclass, occupation, native-country')
plt.show()
```

It is clear that the majority of missing values occur for low income individuals. The amount is missingness is higher than expected. If we assumed the missingness to be occurring for the same reason, then we would expect $37155/11687 \times 750 = 2384$ missing values for low income individuals. This is much lower than the actual value of 5715. This suggests that the data is **not missing at random** and the missingness is informative.

We can see that `workclass` and `occupation` share missing values. This makes sense, as if the occupation is unknown, then the work class is unknown. Conversely, if the occupation is known, then the work class is known too. 

From this brief look into the missing values, we conclude that that low-income individuals are more likely to have missing data and that `workclass` missing values occur together with `occupation` missing values. This shows that the missingness is not ignorable. This knowledge is useful for the subsequent modelling.

```python
# Count the number of unique values in 'education-num'
education_num_counts = X['education-num'].value_counts()

# Count the number of unique values in 'education'
education_counts = X['education'].value_counts()

# Display the results
print("Count of each value in 'education-num':\n", education_num_counts)
print("\nCount of each value in 'education':\n", education_counts)
```

It was difficult to gain insights about `capital-gain` and `capital-loss` due to the high number of values equal to 0. To remedy this, We ignored the zero values. We also look at their distributions for low income and high income individuals separately. 

```python
# Filter out rows where 'capital-gain' or 'capital-loss' is 0
X_filtered = X[(X['capital-gain'] > 0) | (X['capital-loss'] > 0)]

# Ensure to set the style of seaborn for better aesthetics
sns.set(style="whitegrid")

# Set up the figure for subplots
plt.figure(figsize=(15, 6))

# Box plot for 'capital-gain'
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
sns.boxplot(data=X_filtered, x=y[y.columns[0]], y='capital-gain', hue=y[y.columns[0]], palette='Set1', legend=False)
plt.title("Box Plot of Capital Gain by Income")
plt.xlabel("Income")
plt.ylabel("Capital Gain")
plt.grid(False)

# Box plot for 'capital-loss'
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
sns.boxplot(data=X_filtered, x=y[y.columns[0]], y='capital-loss', hue=y[y.columns[0]], palette='Set1', legend=False)
plt.title("Box Plot of Capital Loss by Income")
plt.xlabel("Income")
plt.ylabel("Capital Loss")
plt.grid(False)

# Adjust layout to make space for suptitle and avoid overlap
plt.tight_layout()  # Adjust layout to prevent overlap
plt.suptitle("Distributions of Capital Gain and Capital Loss (Ignoring Zero Values) by Income", y=1.03)  # Adjust title position

plt.show()
```

This allowed us to see that the positive capital gains are higher on average for the high income individuals, with several outliers at very high values. The upper quartiles for the capital losses were similar, but the maximum loss was slightly higher for the low income individuals. At a glance, we can see a more significant difference in distributions for capital gain, suggesting that using that variable can more easily help distinguish between low and high income individuals.

We will conclude our EDA by looking at the links between categorical and numerical variables. Investigating all variables would produce a very dense plot, so we will omit some variables for easier interpretation.
